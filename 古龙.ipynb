陆小凤

```
import jieba

txt= open("陆小凤传奇.txt", "r", encoding='utf-8').read()
excludes = {"陆小凤","知道","可是","因为"}
words  = jieba.lcut(txt)
counts = {}
for word in words:
    if len(word) <=3:
       continue
    elif word == "陆小风":
        rword = "陆小凤"
    else:
        rword = word
    counts[rword] = counts.get(rword,0) + 1
# for word in excludes:
#     del counts[word]

items = list(counts.items())
items.sort(key=lambda x:x[1], reverse=True) 
for i in range(100):
    word, count = items[i]
    print ("{0:<10}{1:>5}".format(word, count))
```

```
#only count for 4 words
import jieba

txt= open("陆小凤传奇.txt", "r", encoding='utf-8').read()
excludes = {"陆小凤","知道","可是","因为"}
words  = jieba.lcut(txt)
words2 = []
counts = {}
for word in words:
    if len(word) <=3:
       continue
    elif word == "陆小风":
        rword = "陆小凤"
    else:
        rword = word
    words2.append(rword)
    counts[rword] = counts.get(rword,0) + 1
# for word in excludes:
#     del counts[word]

items = list(counts.items())
items.sort(key=lambda x:x[1], reverse=True) 
for i in range(100):
    word, count = items[i]
    print ("{0:<10}{1:>5}".format(word, count))
```

```
import wordcloud

# 构建词云对象w，设置词云图片宽、高、字体、背景颜色等参数
w = wordcloud.WordCloud(width=400,height=300,background_color='white',scale=3,font_path="/usr/share/fonts/truetype/SimHei.ttf")

txtstring=" ".join(words2)
# 调用词云对象的generate方法，将文本传入
w.generate(txtstring)

# 将生成的词云保存为output2-poem.png图片文件，保存到当前文件夹中
w.to_file('output2-poem.png')
display(Image('output2-poem.png'))
```
